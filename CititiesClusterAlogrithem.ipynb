```python
from sklearn.cluster import KMeans
```

```python
mylist = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(x)
    mylist.append(kmeans.inertia_)
plt.plot(range(1, 20), mylist)
plt.title('The Model')
plt.xlabel('K')
plt.ylabel('The Error')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y_predict = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'red', label = 'Group 1')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'blue', label = 'Group 2')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s = 100, c = 'green', label = 'Group 3')
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Group 4')
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Group 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('CLuster Cities Prediction')
plt.xlabel('Data Of Cities')
plt.ylabel('Cities')
plt.legend()
plt.show()
```
